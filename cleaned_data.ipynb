```python
# pull in our data
# clean our data
# save cleaned data to db
# send data to browser using flask
# build a chart
import os
import pandas as pd
from pymongo import MongoClient

client = MongoClient(os.getenv("MONGO_URI", "mongodb://localhost:27017/"))
db = client.vaccinfo
```

```python
file_one = "./country_vaccinations.csv"
vaccines = pd.read_csv(file_one)
fill_na = vaccines.fillna(0)
cleaned = fill_na[["country", "total_vaccinations","daily_vaccinations","date","daily_vaccinations_per_million","vaccines","source_name"]]
cleaned
```

```python
for row in cleaned.iterrows():
    if(row[1].total_vaccinations >= 1):
        db.info.insert_one({
            "country":row[1].country,
            "total_vacs": row[1].total_vaccinations,
            "daily_vacs":row[1].daily_vaccinations,
            "data": row[1].date,
            "vacs_cap": row[1].daily_vaccinations_per_million,
            "made_by": row[1].vaccines,
            "source": row[1].source_name
        })
```
